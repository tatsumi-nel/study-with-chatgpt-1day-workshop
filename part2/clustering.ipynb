```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# サンプルデータを生成します（ランダムな点）
np.random.seed(0)
X = np.random.rand(100, 2)

# K-meansモデルを作成します
kmeans = KMeans(n_clusters=3)  # クラスタの数を指定
kmeans.fit(X)

# クラスタの中心を取得します
cluster_centers = kmeans.cluster_centers_

# 各データポイントがどのクラスタに所属するかを予測します
labels = kmeans.labels_

# 結果をプロットします
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis")
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c="red", marker="x", s=200)
plt.title("K-means Clustering")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# サンプルデータを生成します（ランダムな点）
np.random.seed(0)
X = np.random.rand(100, 2)

# K-meansモデルを作成します
kmeans = KMeans(n_clusters=3)  # クラスタの数を指定
kmeans.fit(X)

# メッシュグリッドを作成します
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# メッシュ上の各点に対するクラスタを予測します
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 結果をプロットします
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap="viridis")
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    c="red",
    marker="x",
    s=200,
)
plt.contourf(xx, yy, Z, alpha=0.3, cmap="viridis")
plt.title("K-means Clustering with Boundaries")
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

# サンプルデータを生成します（ランダムな点）
np.random.seed(0)
X = np.random.rand(100, 2)

# K-meansモデルを作成します
kmeans = KMeans(n_clusters=3)  # クラスタの数を指定
kmeans.fit(X)

# 判定したい点を作成します
points_to_predict = np.array([[0.2, 0.1], [0.4, 0.8], [0.9, 0.5]])

# 判定したい点がどのクラスタに属するか予測します
predicted_labels = kmeans.predict(points_to_predict)

# 各点の所属クラスタを表示します
for i, point in enumerate(points_to_predict):
    print(f"Point {point} belongs to Cluster {predicted_labels[i]}")
```
