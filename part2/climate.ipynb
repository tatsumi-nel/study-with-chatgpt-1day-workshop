```python
import pandas as pd  
import matplotlib.pyplot as plt  
import japanize_matplotlib  

# データの読み込み  
df = pd.read_csv('tokyo.csv')  

# 日付をdatetime型に変換  
df['年月日'] = pd.to_datetime(df['年月日'])  

# グラフの描画  
plt.figure(figsize=(10, 5))  
plt.plot(df['年月日'], df['平均気温(℃)'])  
plt.title('東京の平均気温')  
plt.xlabel('日付')  
plt.ylabel('気温(℃)')  
plt.grid(True)  
plt.show()
```
