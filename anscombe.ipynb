```python
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# JSONファイルを読み込む
with open('sample_data/anscombe.json', 'r') as file:
    data = json.load(file)

# データを4つのシリーズに分割する
series_I = [item for item in data if item['Series'] == 'I']
series_II = [item for item in data if item['Series'] == 'II']
series_III = [item for item in data if item['Series'] == 'III']
series_IV = [item for item in data if item['Series'] == 'IV']

# グラフを作成
plt.figure(figsize=(12, 8))

# 各シリーズに対して線形回帰分析を行い、回帰直線を作成
def perform_linear_regression(series, series_name):
    X = np.array([item['X'] for item in series]).reshape(-1, 1)
    Y = np.array([item['Y'] for item in series])

    # 線形回帰モデルを作成
    model = LinearRegression()
    model.fit(X, Y)

    # 回帰直線の係数と切片を取得
    slope = model.coef_[0]
    intercept = model.intercept_

    # 回帰直線を描画
    plt.scatter(X, Y, label=f'{series_name} Data')
    plt.plot(X, model.predict(X), label=f'{series_name} Regression Line', color='red')

    # 回帰直線の関数形を表示
    print(f'{series_name} Regression Line: Y = {slope:.2f}X + {intercept:.2f}')

# 各シリーズに対して線形回帰分析を実行し、回帰直線を作成
perform_linear_regression(series_I, 'Series I')
perform_linear_regression(series_II, 'Series II')
perform_linear_regression(series_III, 'Series III')
perform_linear_regression(series_IV, 'Series IV')

# グラフの設定
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Analysis')
plt.grid(True)

# グラフを表示
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# JSONファイルを読み込む
with open('sample_data/anscombe.json', 'r') as file:
    data = json.load(file)

# データを4つのシリーズに分割する
series_I = [item for item in data if item['Series'] == 'I']
series_II = [item for item in data if item['Series'] == 'II']
series_III = [item for item in data if item['Series'] == 'III']
series_IV = [item for item in data if item['Series'] == 'IV']

# グラフを4つまとめて表示
plt.figure(figsize=(12, 8))

# 各シリーズに対して線形回帰分析を行い、データと回帰直線をプロット
def plot_linear_regression(series, series_name, subplot_position):
    X = np.array([item['X'] for item in series]).reshape(-1, 1)
    Y = np.array([item['Y'] for item in series])

    # 線形回帰モデルを作成
    model = LinearRegression()
    model.fit(X, Y)

    # 回帰直線の係数と切片を取得
    slope = model.coef_[0]
    intercept = model.intercept_

    # サブプロットを作成
    plt.subplot(2, 2, subplot_position)

    # Xの範囲を(0, 20)に制限
    X_range = np.linspace(0, 20, 100).reshape(-1, 1)

    # データと回帰直線をプロット
    plt.scatter(X, Y, label=f'{series_name} Data')
    plt.plot(X_range, model.predict(X_range), label=f'{series_name} Regression Line', color='red')

    # グラフの設定
    plt.title(f'{series_name} - Regression Line: Y = {slope:.2f}X + {intercept:.2f}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.xlim(0, 20)  # 横軸の範囲を設定
    plt.ylim(0, 16)  # 縦軸の範囲を設定
    plt.legend()
    plt.grid(True)

# 各シリーズに対して線形回帰分析を実行し、データと回帰直線をプロット
plot_linear_regression(series_I, 'Series I', 1)
plot_linear_regression(series_II, 'Series II', 2)
plot_linear_regression(series_III, 'Series III', 3)
plot_linear_regression(series_IV, 'Series IV', 4)

# グラフを表示
plt.tight_layout()
plt.show()


```
